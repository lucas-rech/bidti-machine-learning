```python
from sklearn.ensemble import RandomForestClassifier

import pickle

from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import classification_report
```

```python
with open('../data/student_perfomance.pkl', 'rb') as f:
    X_students_train, y_students_train, X_students_test, y_students_test = pickle.load(f)
```

```python
random_forest = RandomForestClassifier(n_estimators = 15, criterion = 'entropy', random_state = 0)
random_forest.fit(X_students_train, y_students_train)
```

```python
predictions = random_forest.predict(X_students_test)
predictions
```

```python
y_students_test
```

```python
cm = ConfusionMatrix(random_forest)
cm.fit(X_students_train, y_students_train)
cm.score(X_students_test, y_students_test)
```

```python
print(classification_report(y_students_test, predictions))
```
