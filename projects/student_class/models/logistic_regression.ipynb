```python
from sklearn.linear_model import LogisticRegression
import pickle

from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import classification_report
```

```python
with open('../data/student_perfomance.pkl', 'rb') as f:
    X_students_train, y_students_train, X_students_test, y_students_test = pickle.load(f)
```

```python
logistic_regression = LogisticRegression(random_state = 1)
logistic_regression.fit(X_students_train, y_students_train)
```

```python
predictions = logistic_regression.predict(X_students_test)
predictions
```

```python
cm = ConfusionMatrix(logistic_regression)
cm.fit(X_students_train, y_students_train)
cm.score(X_students_test, y_students_test)
```

```python
print(classification_report(y_students_test, predictions))
```
